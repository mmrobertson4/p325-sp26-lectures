## Matrix Math

So far, we've only talked about the standard Python operators (`+`, `-`, `*`, `/`, and `**`) that are applied element-wise to NumPy arrays. To be useful for scientifc computing, arrays must also handle operations like matrix multiplication. The following section will introduce more advanced matrix operations using Python.

### Matrix Multiplication

To multiply two matrices together in Python, you can use the `@` operator. For example:

```python
import numpy as np

matrixa = np.zeros((2,2)) + 1 # fill a 2x2 matrix with 1's
matrixb = np.zeros((2,2)) + 2 # fill a 2x2 matrix with 2's

print(matrixa @ matrixb)
```

First, we defined a matrix filled with zeros and used the element-wise operator `+` to add `1` to each matrix element. We did the same for the next matrix, but instead added `2`. Finally, we performed the matrix multipication of the two using the `@` operator.

The `@` operator was introduced in Python 3.5. If you're looking at code that was written earlier than this, you will instead see the function `np.matmul()`. In general, the `@` operator provides a cleaner, more readable expression. For example, consider multiplying three objects together:

```python
print(matrixa @ matrixa @ matrixa)
print(np.matmul(matrixa, np.matmul(matrixa, matrixa)))
```

### Multiplying a Matrix by a Vector

Similarly, you can multiply a matrix (2-D array) by a vector (1-D array).

```python
veca = np.ones(2)

print(matrixa @ veca)
```

Because a 1-D array doesn't have any conception of being a row or column vector, the order doesn't matter:

```python
print(veca @ matrixa)
```

Internally, when a 2-D array is multiplied by a 1-D array, the 1-D array is promoted to a 2-D array:
* If the first argument is 1-D, it is promoted to a matrix by prepending a 1 to its dimensions.
* If the second argument is 1-D, it is promoted to a matrix by appending a 1 to its dimensions.

### Other Convenient Attributes and Functions

In the previous section you were asked to manually make the identity matrix. There is also a NumPy function for accomplishing this:

```python
ident = np.eye(4)
print(ident)
```

You can also calculate the *trace* of a matrix, i.e., the sum of its diagonal elements:

```python
print(np.trace(ident))
```

Finally, ever NumPy array includes the `T` attribute, which contains the transpose of the array:

```python
mat = np.linspace(1,25,25).reshape(5,5)
print(mat, '\n')
print(mat.T)
```

### Linear Algebra Operations

For more advanced operations, NumPy offers the `linalg` module. This allows you to calculate the **inverse** of a square, non-singular matrix, the **determinant** of a square matrix, and the **eigen values** of a matrix, among other things.

Let's start by calculating the determinant of a matrix:

```python
mat = np.array([[4, 7], 
                [2, 6]])
print(np.linalg.det(mat))
```

Next, we'll calculate the inverse and check that it works as expected:

```python
mat = np.array([[4, 7], 
                [2, 6]])
inv = np.linalg.inv(mat)
print(inv, '\n')
print(mat @ inv)

```

Finally, let's print the eigen values:

```python
eigvals, eigvecs = np.linalg.eig(mat)
print(eigvals)
print(eigvecs)
```

## Practice

Imagine you are a biologist studying a species with two life stages: **Juveniles** and **Adults**.
1. **Survival:** Only 50% of Juveniles survive to become Adults. 80% of Adults survive each year.
2. **Reproduction:** Juveniles don't reproduce. Each Adult produces 2 new Juveniles per year.
We can represent the population at any time $t$ as a vector $\mathbf{v}_t$ and the transition rules as a matrix $M$.

To find the population for the next year, we use the formula:
$$
\mathbf{v}_{t+1} = M \mathbf{v}_t
$$
Where:
* $M = \begin{pmatrix} 0 & 2 \\ 0.5 & 0.8 \end{pmatrix}$
* $\mathbf{v}_t = \begin{pmatrix} \text{Juveniles} \\ \text{Adults} \end{pmatrix}$

Start by defining the matrices $M$ and $v_t$.

```python
M = np.array([[0,2], [0.5,0.8]])
v = np.array([100, 100]).reshape(2,1)

print(M)
print(v)
```

Next, perform the matrix multiplication: $M \mathbf{v}_t$.

```python
mult = M @ v

print(mult)
```

What is the juvenile population after 1 year? Can you calculate it after 10 years?

```python
M = np.array([[0,2], [0.5,0.8]])
v = np.array([100, 100]).reshape(2,1)

i = 0
while i < 10:
    i += 1
    v_next = M @ v
    v = v_next
    print(f'After {i} years:\n {v}')

```
