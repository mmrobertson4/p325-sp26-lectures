# User Input and While Loops

## Practice

When buying tickets to a movie, your age determines which movies you're allowed to see (G, PG, PG-13, R). Write a `while` loop that asks a user for their age, then prints the highest rated movie they are allowed to see. Include a quit prompt that will use `break` to stop the loop. Can you think of a reasonable use for the `continue` statement in this example?

```python
question = "Enter your age (or q to quit):  "

while True:
    response = input(question)
    if response == "q":
        break

    age = int(response)

    if age < 5:
        print("You can see a G movie")
    elif 5 <= age < 13:
        print("You can see G, PG")
    elif 13 <= age < 18:
        print("You can see G, PG, PG-13")
    else:
        print("You can see any movie")

    
```
